### Final Project

#### Scope

The scope focuses on the **pipeline of data science** - exploring, preprocessing, splitting, training, validating. Accuracy may not be the sole focus, but usually a good pipeline will provide relatively decent performance.

#### Topics

Announced in google classroom.

#### Deliverables

Deliverables will be focused on boosting your resume, toward achieving a good career.

- Your python (.py)/notebook (.ipynb) file

- A medium article that you are proud of

- A github link, together with a useful README

- A presentation (PPT/PDF)

#### Group number

1-2 people.  This is to make sure contribution is recognized.


#### Criteria

START EARLY!

Projects will be evaluated based on:

- Creativity (40%)
- Critical Thinking, Scientific Thinking, Depth, Rigor (40%)
- Element of surprise (10%) - surprise findings, surprising interpretation, surprising new techniques, etc.
- Presentation and communication (10%)

#### Presentation

15 minutes to present your project.  Make sure to cover:

- Motivation (what is the project about)
- Experiments (describe what you compare, how you measure)
  - we want to see you use some experiment tool
- Discussion  (did you get what you expect, what you have learn)
- Future Work  (what can you do better)
- Demo  (show some demo)
